#Fault diagnosis model for early warnings in electrical transformers.

Efficient management of electrical transformers is essential to ensure the reliability of electrical networks, but the lack of early warnings can lead to unscheduled events and increased repair costs. In this context, it is proposed to develop a fault diagnosis model for electrical transformers based on the analysis of electrical variables and data processing techniques. This model aims to generate early alerts that improve operational efficiency, prolong the useful life of equipment and contribute to the advancement of fault diagnosis systems in electrical networks. The expected results have a significant impact on the reliability of electrical networks, maintenance efficiency, stability of electrical systems and a reduction in operating and repair costs, with both social, economic and environmental benefits.

The detection and diagnosis of faults in electrical transformers has been a topic of interest since the beginning of electrification, and has acquired critical importance in the global electrical infrastructure. The various methods of fault detection and signal analysis applied to electrical transformers are of essential importance for research; detection and diagnosis techniques that are currently used to detect faults that can occur in transformers, which are mainly based on the most common faults of a transformer including overheating, short circuits or faults in the winding, insulation problems and partial discharges, among others (Youssef et al., 2022).

According to (Juan Z et al., 2019), among the most common failure detection methods are: i) monitoring of gases dissolved in the insulating oil, which analyzes the gases generated by failures such as hydrogen (H2), methane (CH4), acetylene (C2 H2), ethylene (C2 H4), ethane (C2 H6), carbon monoxide (CO) and carbon dioxide (CO2) to identify incipient problems in the transformer (Lynn Hamrick, 2009); ii) infrared thermography, which detects temperature variations that may indicate overheating or contact problems in the active part of the transformer; iii) measuring electrical parameters such as current and voltage to identify imbalances or short circuits; and iv) the use of diagnostic systems based on signal processing and machine learning techniques, which can analyze complex data to detect failure patterns. Each of these methods has its advantages and limitations, and the choice of the appropriate technique depends on factors such as the precision required, the type of faults to be detected and the operating conditions of the transformer. Integrating multiple detection approaches can provide a more complete assessment of transformer health and improve the reliability of fault detection.

#Database call
The script defines a function called $impDB$ that takes a DataBase argument, which is expected to be the name of the database file to load. Within the function, the path of the file in the Google Colab environment which was cloned from the Git repository is constructed, and then the file is attempted to be opened. If the file opens correctly, each line is processed to convert the values to float numbers if possible, keeping the non-numeric values as they are, and stored in a $DB \quad list$. This list of lists represents the structure of the database loaded from the file. If any error occurs during this process, a message is printed indicating that the file does not exist in the specified path and the program ends.

```python
import time
def impDB(DataBase): # O(1)
    try:  # O(1)
        db_path = f'/content/Proyecto1/DatosMonitoringTransformer.txt'  # O(1)
        with open(db_path) as f:  # O(1)
            DB = []  # O(1)
            for line in f: # O(n)
                row = [float(value) if value.replace('.', '', 1).isdigit() else value for value in line.strip().split(',')] # O(n^2)
                DB.append(row) # O(1)
        return(DB) # O(1)
    except: # O(1)
        print("The .txt file does not exist in the specified path") # O(1)
        quit() # O(1)
```

#Elimination of zero currents
The code defines the function called deleteZeros which takes two arguments: Database, which will be our database, and Columns, which is a list of column indexes. The function checks whether DataBase is really a list and, if not, generates a TypeError exception. Then, create a new empty $DBZ$ list and then each row will be traversed in the database. If all columns specified in Columns have values equal to zero in a row, that row is ignored; otherwise, it is added to the $DBZ \quad list$. Finally, the function returns the $DBZ \quad list$, which contains the Database rows that do not have zeros in the specified columns.

Inputs: The function takes two parameters: DataBase, which is our database, and Columns, which is a list with the number of columns to be filtered.

Output: Returns a modified list $(DBZ)$ containing the DataBase rows that do not have zeros in the specified columns.

Operation: Checks if the Database is a list. If not, a TypeError error is generated. Creates a new $DBZ \quad list$ using list understanding to filter the database rows that do not have zeros in the columns specified in Columns and returns the filtered $DBZ \quad list$.

```python
#A. Deficient
def deleteZeros(DataBase,Columns): # O(1)
    try: # O(1)
        if not isinstance(DataBase,list): # O(1)
            raise TypeError("The input of the function to remove zeros from the current column is not an array")  #O(1)
    except TypeError as Error: # O(1)
        print(Error)  # O(1)
    DBZ=[]  # O(1)
    for row in DataBase: # O(n)
        if all(row[Column] == 0 for Column in Columns): # O(n)
            continue # O(1)
        DBZ.append(row) # O(1)
    return(DBZ) # O(1)

#B. Efficient
def deleteZerosOP(DataBase, Columns):  # O(1)
    if not isinstance(DataBase, list):  # O(1)
        raise TypeError("The input of the function to remove zeros from the current column is not an array")  # O(1)
    try:  # O(1)
        DBZ = [row for row in DataBase if not all(row[column] == 0 for column in Columns)]  # O(n)
    except TypeError as Error:  # O(1)
        print(Error)  # O(1)
    return DBZ  # O(1)
```

Complexity

A. Deficient:

Time complexity: $O(1) + O(1) + O(1) + O(1) + O(1) + O(n) * ( O(n) + O(1) + O(1)) + O(1) = O(n ^ 2)$

Space complexity: $ O(n ^ 2)$ (for the DBZ list)

So the overall time complexity of the deleteZeros function is $O(n^2)$, determined by the number of rows in the database and the number of columns to check in each row.

B. Efficient:

Time complexity: $O(1) + O(1)+ O(1) + O(1) + O(n) + O(1) + O(1) = O(n)$

Space complexity: $ O(n)$ (for the DBZ list)

So, the overall time complexity of the deleteZeros function after optimizing the code we can notice that it improved and is now $O(n)$, determined by the number of rows in the database and the number of columns to check in each row.

Optimization

The changes we made to try to optimize this function were the following:
- Use of List Comprehension: Instead of using an explicit for loop to filter the database rows, a list comprehension was used. This is more efficient and compact.
- Conditional optimization: The conditional if $all(row[column] == 0\quad for \quad column \quad in \quad Columns)$ checks if all elements of a row are zero in the specified columns. This is done most efficiently by using all directly in the list comprehension.


These changes reduce the execution time of the algorithm by having fewer operations, thus avoiding unnecessary repetition of them, and also by performing critical checks before the main loop.

```python
column = [4, 5, 6]
print("Database size with zeros: " + str(len(DB)) + '\n')
st = time.time()
DBZ = deleteZeros(DB,column)
et = time.time()
print("Script A ")
print("Database size without zeros: " + str(len(DBZ)))
print(f'Elapsed time: {(et-st)*1000} ms\n')
print("Script B ")
st = time.time()
DBZ = deleteZerosOP(DB,column)
et = time.time()
print("Database size without zeros: " + str(len(DBZ)))
print(f'Elapsed time: {(et-st)*1000} ms')
```

# Power calculation
In this script we work on the function called $calculate\_power\_per\_record$ that calculates the approximate power per record in our database. For each record in the database (excluding the header), the function calculates the average phase voltage, line voltage, average current and finally the power per record using the three-phase power formulas. The results are rounded to two decimal places and stored in a list that is returned at the end. This approach allows obtaining an estimate of the energy consumed by each record in the database, which can be useful for subsequent analysis or specific calculations related to energy consumption.

Inputs: The $calculate\_power\_per\_recordOP1$ function takes a list of records (database) as input.

Output: Returns a list of approximate powers calculated from the records.

How it works: Iterates over the database records, starting from the second record $(database[1:])$. For each record, calculate the approximate power using the electrical power formulas and save each approximate power in the power list.

EQUATIONS

- Average phase-to-phase voltage
$$
V_L = \sqrt{3} \times \left( \frac{{\sum V_f}}{N} \right)
$$

Where:

$V_L:$ Phase-to-phase voltage.

$V_f:$ Phase voltage.

$N:$ Number of records.

\\

- Average current

$$
I_{avg} = \frac{{\sum I}}{N}
$$

Where:

$I_{avg}$: Average current.

$I$: Current.

$N$: Number of records.

\\

- Power

$$
P = \frac{{\sqrt{3} \times V_L \times I_{avg}}}{1000}
$$

Where:

$P$: Approximate power.

$V_L$: Phase-to-phase voltage.

$I_{avg}$: Average current.

```python
#A. Deficient
import math
def calculate_power_per_record(database): # O(1)
    powers = [] # O(1)
    for register in database[1:]: # O(n)
        phase_voltages = register[1:4]  # O(1)
        currents = register[4:7]  # O(1)

        vf_average = sum(phase_voltages) / len(phase_voltages) # O(1)
        I_average = sum(currents) / len(currents) # O(1)
        vl = math.sqrt(3) * vf_average # O(1)

        power_registration = (math.sqrt(3) * vl * I_average)/1000 # O(1)
        approximate_power = round(power_registration, 2)  # O(1)

        powers.append(approximate_power)  # O(1)

    return (powers) # O(1)

#B. Efficient
def calculate_power_per_recordOP(database): # O(1)
    powers = [] # O(1)
    for register in database[1:]: # O(n)
        powers.append (round((math.sqrt(3) * (math.sqrt(3) * (sum(register[1:4]) / 3)) * (sum(register[4:7]) / 3)) / 1000, 2)) # O(1)
    return powers # O(1)

```

Complexity

A.  Deficient
- Time complexity: $$O(1) + O(1) + O(n) * (O(1) + O(1) + O(1) + O(1) + O(1) + O(1) + O(1) + O(1) + O(1) + O(1) + O(1) + O(1)) + O(1)$ = $ O(n) $$

Iterate over records: $O(n)$, where n is the number of records in the database and mathematical operations and power calculations per record: $O(1)$ for each record.

- Space complexity: $O(n) (for \quad  the \quad  powers \quad  list)$ where n is the number of records in the database, since we store the calculated power for each record.


B. Efficient
- Time complexity: $O(1) + O(1) + O(n) * O(1) + O(1) =  O(n) $

- Space complexity: $O(n) (for \quad  the \quad  powers \quad  list)$, where n is the number of records in the database, since we store the calculated power for each record in the list.

The time complexity of the code is dominated by the iteration over records in the database, which has a complexity of $O(n)$, where n is the number of records. The space complexity is $O(n)$ due to the creation of the powers list that stores the power calculated for each record.

Optimization
Reduction of lines of code: We combined the power calculation and rounding and all the operations that depended on it in a single line to simplify the code and reduce the number of operations and therefore make the code more efficient in terms of time.

#Calculation of the standard deviation
In this script we work on the function called standard_deviation which calculates the standard deviation of three current data sets (I1, I2 and I3) in our database. First, the code collects the records from each set into separate lists. Second, the average of each set of currents is calculated. Third, the average distance of each current from its respective mean is calculated by squaring and summing the results. Fourth, the square root of the mean of these distances is calculated to obtain the standard deviation of each set of currents. Fifth and finally, we proceed to calculate the upper and lower limits for each set of currents by adding and subtracting the standard deviation from the respective mean. The function returns the standard deviations and the upper and lower limits for each set of currents. This approach allows me to analyze the variability of the current current data in the database and establish limits to identify outliers or unusual values later as the development of the thesis work progresses.

Inputs: The standard_deviationOP function takes a list of records $(database)$ as input.

Output: Returns the values of the standard deviations $(DEI1, DEI2, DEI3)$ and the upper and lower limits for each current $(LIM1I1, LIM2I1, LIM1I2, LIM2I2, LIM1I3, LIM2I3).$

Functioning:
Current Sums:
$I1_sum, I2_sum, I3_sum$: The currents in each record are added together to calculate averages later.

EQUATIONS

Current Averages:
$I1avg, I2avg, I3avg$: They are calculated by dividing the sum of the currents by the number of registers minus one $(n-1)$, assuming that the first register is omitted.

$$\text{I 1avg} = \frac{\text{sum I1}}{\text{I1(n)}}$$

Sums of Average Distances:
$distMediaI1_sum, distMediaI2_sum, distMediaI3_sum:$ The mean squared distances of each stream from the average are calculated and added to calculate standard deviations later.

$$\text{DistMediaI1} = \text({I 1(n)} - \text{I 1avg})^2$$

Standard Deviations:
$DEI1, DEI2, DEI3:$ They are calculated by dividing the sum of the squared mean distances by n-1 and then taking the square root.

$$\text{DE1} = \sqrt{\frac{\sum (\text{DistMediaI1)}}{DistMediaI1(n)}}$$

Upper and Lower Limits:
$LIM1I1, LIM2I1, LIM1I2, LIM2I2, LIM1I3, LIM2I3: $ They are calculated by adding or subtracting the standard deviation from the corresponding averages.


$$\text{Lmite Superior} = \text{I 1avg} + \text{DE1} \\
\text{Lmite Inferior} = \text{I 1avg} - \text{DE1}$$

```python
import math
import time
#A. Deficient
def standard_deviation(database): # O(1)
    I1 = [] # O(1)
    I2 = [] # O(1)
    I3 = [] # O(1)
    for row in database[1:]: # O(n)
        I1.append(row[4]) # O(1)
        I2.append(row[5]) # O(1)
        I3.append(row[6]) # O(1)

    I1avg = sum(I1) / len(I1) # O(1)
    I2avg = sum(I2) / len(I2) # O(1)
    I3avg = sum(I3) / len(I3) # O(1)

    distMediaI1 =[] # O(1)
    distMediaI2 =[] # O(1)
    distMediaI3 =[] # O(1)

    for row in database[1:]: # O(n)
        distMediaI1.append(abs(row[4]-I1avg)**2) # O(1)
        distMediaI2.append(abs(row[5]-I2avg)**2) # O(1)
        distMediaI3.append(abs(row[6]-I3avg)**2) # O(1)

    DEI1 = round(math.sqrt(sum(distMediaI1) / len(distMediaI1)),2) # O(1)
    DEI2 = round(math.sqrt(sum(distMediaI2) / len(distMediaI2)),2) # O(1)
    DEI3 = round(math.sqrt(sum(distMediaI3) / len(distMediaI3)),2) # O(1)

    LIM1I1 = I1avg + DEI1 # O(1)
    LIM2I1 = I1avg - DEI1 # O(1)
    LIM1I2 = I2avg + DEI2 # O(1)
    LIM2I2 = I2avg - DEI2 # O(1)
    LIM1I3 = I3avg + DEI3 # O(1)
    LIM2I3 = I3avg - DEI3 # O(1)

    return(DEI1,DEI2,DEI3,LIM1I1,LIM2I1,LIM1I2,LIM2I2,LIM1I3,LIM2I3) # O(1)

#B. Efficient
def standard_deviationOP(database):
    n = len(database)  # O(1)
    if n <= 1:  # O(1)
        return 0, 0, 0, 0, 0, 0, 0, 0, 0  # O(1)

    I1_sum,I2_sum,I3_sum,distMediaI1_sum,distMediaI2_sum,distMediaI3_sum  = 0,0,0,0,0,0  # O(1)

    for row in database[1:]:  # O(n)
        I1_sum += row[4]  # O(1)
        I2_sum += row[5]  # O(1)
        I3_sum += row[6]  # O(1)

    I1avg = I1_sum / (n - 1)  # O(1)
    I2avg = I2_sum / (n - 1)  # O(1)
    I3avg = I3_sum / (n - 1)  # O(1)

    for row in database[1:]:  # O(n)
        distMediaI1_sum += (row[4] - I1avg) ** 2  # O(1)
        distMediaI2_sum += (row[5] - I2avg) ** 2  # O(1)
        distMediaI3_sum += (row[6] - I3avg) ** 2  # O(1)

    DEI1 = round(math.sqrt(distMediaI1_sum / (n - 1)), 2)  # O(1)
    DEI2 = round(math.sqrt(distMediaI2_sum / (n - 1)), 2)  # O(1)
    DEI3 = round(math.sqrt(distMediaI3_sum / (n - 1)), 2)  # O(1)

    LIM1I1 = I1avg + DEI1  # O(1)
    LIM2I1 = I1avg - DEI1  # O(1)
    LIM1I2 = I2avg + DEI2  # O(1)
    LIM2I2 = I2avg - DEI2  # O(1)
    LIM1I3 = I3avg + DEI3  # O(1)
    LIM2I3 = I3avg - DEI3  # O(1)

    return DEI1, DEI2, DEI3, LIM1I1, LIM2I1, LIM1I2, LIM2I2, LIM1I3, LIM2I3  # O(1)

#C. Deficient with numpy
import numpy as np

def standard_deviationOP1(database): # O(1)
    currents = np.array(database[1:])[:, 4:7].astype(float) # O(n^2)

    current_avgs = np.mean(currents, axis=0) # O(n)
    current_stds = np.std(currents, axis=0) # O(n)

    limits = np.column_stack((current_avgs + current_stds, current_avgs - current_stds)) # O(n)

    std_devs = tuple(np.round(current_stds, decimals=2)) # O(n)
    limits_rounded = tuple(np.round(limits, decimals=2).flatten()) # O(n)

    return std_devs + limits_rounded # O(1)

```

- Time complexity: $ O(1) + O(n^2) + O(n) + O(n) + O(n) + O(n) + O(n) + O(1) = O(n^2)$

- Space complexity: $O(1)$

In summary, the overall time complexity of the function standard_deviationOP1 is $O(n^2)$ and with this in mind the script is much poorer using numpy, where n is the number of rows in the database minus 1.

Optimization

Script B gains efficiency in memory usage: In Script B, variables are used to accumulate the sums (I1_sum, I2_sum, I3_sum) and the squared differences (distMediaI1_sum, distMediaI2_sum, distMediaI3_sum) without needing to create additional lists as in Script A. This reduces the amount of memory used, especially in large databases, where lists can take up a significant amount of memory.

# Object-oriented programming (OOP)

```python
class DataBaseCheck:
  def __init__(self,DB:list):
    self.__DB__ = DB
    self.__boolean__ = self.__check__()

  def __check__(self):
    if isinstance(self.__DB__,list):
      return True
    else:
      return False

class DelZeros(DataBaseCheck):
  def __init__(self,DB:list,Columns:list):
    super().__init__(DB)
    self.__Columns__ = Columns
    st = time.time()
    self.__DBZ__ = self.__deleteZeros__()
    et = time.time()
    print(f'Elapsed execution time for DeleteZeros: {(et-st)*1000} ms\n')

  def __deleteZeros__(self):
    try:
      if not self.__boolean__:
        raise TypeError("The database must be List type")
    except TypeError as Error:
      print(Error)

    Temp_DBZ = [row for row in self.__DB__ if not all(row[column] == 0 for column in self.__Columns__)]
    return(Temp_DBZ)

  def getDBZ(self):
    return self.__DBZ__

class PowerCalculation(DataBaseCheck):
  def __init__(self,DB:list):
    super().__init__(DB)
    st = time.time()
    self.__Powers__ = self.__calculate_power_per_record__()
    et = time.time()
    print(f'Elapsed execution time for PowerCalculation: {(et-st)*1000} ms\n')
  def __calculate_power_per_record__(self):
    try:
      if not self.__boolean__:
        raise TypeError("The database must be List type")
    except TypeError as Error:
      print(Error)
    Powers=[]
    for register in self.__DB__[1:]:
        phase_voltages = register[1:4]
        currents = register[4:7]

        sum_phase_voltages = sum(phase_voltages)
        len_phase_voltages = len(phase_voltages)
        sum_currents = sum(currents)
        len_currents = len(currents)

        vf_average = sum_phase_voltages / len_phase_voltages
        vl = math.sqrt(3) * vf_average
        I_average = sum_currents / len_currents

        power_registration = (math.sqrt(3) * vl * I_average) / 1000
        approximate_power = round(power_registration, 2)
        Powers.append(approximate_power)

    return Powers

  def getPowers(self):
    return self.__Powers__

class StandardDeviaton(DataBaseCheck):
    def __init__(self,DB):
        super().__init__(DB)
        self.__n__ = len(self.__DB__[1:])
        st = time.time()
        self.__avg__ = self.__average__()
        self.__SD__ = self.__standadDeviaton__()
        self.__Limits__ = self.__limits__()
        et = time.time()
        print(f'Elapsed execution time for StandardDeviaton : {(et-st)*1000} ms\n')
    def __average__(self):
        try:
            if not self.__boolean__:
                raise TypeError("The database must be List type")
        except TypeError as Error:
            print(Error)

        if self.__n__ <= 1:
            return []
        I1_sum = 0
        I2_sum = 0
        I3_sum = 0

        for row in self.__DB__[1:]:
            I1_sum += row[4]
            I2_sum += row[5]
            I3_sum += row[6]

        return [I1_sum / self.__n__, I2_sum / self.__n__, I3_sum / self.__n__]

    def __standadDeviaton__(self):
        try:
            if not self.__boolean__:
                raise TypeError("The database must be List type")
        except TypeError as Error:
            print(Error)

        DistMediaI1_sum = 0
        DistMediaI2_sum = 0
        DistMediaI3_sum = 0

        for row in self.__DB__[1:]:
            DistMediaI1_sum +=(row[4] - self.__avg__[0]) ** 2
            DistMediaI2_sum +=(row[5] - self.__avg__[1]) ** 2
            DistMediaI3_sum +=(row[6] - self.__avg__[2]) ** 2

        return [round(math.sqrt(DistMediaI1_sum / self.__n__),2), round(math.sqrt(DistMediaI2_sum / self.__n__),2), round(math.sqrt(DistMediaI3_sum / self.__n__),2)]

    def __limits__(self):
        return [[self.__avg__[0] + self.__SD__[0], self.__avg__[0] - self.__SD__[0]],[self.__avg__[1] + self.__SD__[1], self.__avg__[1] - self.__SD__[1]],[self.__avg__[2] + self.__SD__[2],self.__avg__[2] - self.__SD__[2]]]

    def getSD(self):
        return self.__SD__

    def getavg(self):
        return self.__avg__

    def getLimits(self):
        return self.__Limits__

class DataBaseHandling:
    def __init__(self,path:str,I1=4,I2=5,I3=6):
        self.__Path__ = path
        self.__I_Columns__ = [I1,I2,I3]
        self.__DB__ = self.__importDB__(self.__Path__)
        self.__DBZ__ = DelZeros(self.__DB__,self.__I_Columns__)
        self.__PW__ = PowerCalculation(self.__DBZ__.getDBZ())
        self.Power = []
        self.__SD__ = StandardDeviaton(self.__DBZ__.getDBZ())

    def __importDB__(self,path:str):
        try:
            fhand = open(path)
        except:
            print("The .txt file does not exist in the specified path")
            quit()

        DB = []
        for line in fhand:
            row = [float(value) if value.replace('.', '', 1).isdigit() else value for value in line.strip().split(',')]
            DB.append(row)
        return(DB)

    def getDB (self):
        return self.__DB__

    def getDBZ (self):
        return self.__DBZ__.getDBZ()

    def getPowers (self):
        return self.__PW__.getPowers()

    def getSD (self,index:int=0):
      if index == 0:
        return self.__SD__.getSD()
      elif index == 1:
        return self.__SD__.getLimits()
      elif index == 2:
        return self.__SD__.getavg()

Base_de_Datos = DataBaseHandling("DatosMonitoringTransformer.txt")
print("Script for the Database call")
print(f"Have been loaded {len(Base_de_Datos.getDB())} records.\n")
print("Script for the DeleteZeros")
print(f"Database size without zeros:  {len(Base_de_Datos.getDBZ())}\n")
print("Script for the PowerCalculation")
print("Primeros 30 valores de potencias:")
for i, (register, power) in enumerate(zip(Base_de_Datos.getDBZ()[1:30], Base_de_Datos.getPowers()[:30]), start=1):
    print(f"Register {i}: Data = {register}, Power = {power} KvA")
print("\n")
DE1, DE2, DE3 = Base_de_Datos.getSD()
print(f"Standard deviation I1: {DE1}")
print(f"Standard deviation I2: {DE2}")
print(f"Standard deviation I3: {DE3}\n")

limits = Base_de_Datos.getSD(1)
print(f"Upper limit I1: {limits[0][0]}")
print(f"Lower limit I1: {limits[0][1]}")
print(f"Upper limit I2: {limits[1][0]}")
print(f"Lower limit I2: {limits[1][1]}")
print(f"Upper limit I3: {limits[2][0]}")
print(f"Lower limit I3: {limits[2][1]}\n")
```

#Conclusions
With the results obtained, we can conclude the following for each of our scripts:

1. DeleteZeros: The difference in execution times between normal programming and object-oriented programming is minimal. Both implementations have similar times, suggesting that the complexity of this process is not significantly affected by the type of programming used.

2. PowerCalculation: Here we see a more noticeable difference in execution times between normal and object-oriented programming. The object-oriented programming version seems to be a little slower in this case. This could indicate that the way data is processed in the object-oriented implementation introduces some overhead compared to the other implementation.

3. StandardDeviaton: As with DeleteZeros, the execution times are quite similar between both implementations. This suggests that the standard deviation calculation logic, at least in this particular implementation, is not significantly affected by the change in scheduling type.

In general, these results show us that the impact of the type of programming $(normal\quad vs. \quad object-oriented)$ on execution times depends on the complexity and specific structure of each process. In some cases, there may be noticeable differences, while in other cases, the differences are minimal or non-existent.
